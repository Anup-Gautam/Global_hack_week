```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4]
y = [6,7,8,9]
plt.plot(x,y)
plt.show()
plt.plot(x,np.power(y,2))
plt.show()
```

```python
x = np.linspace(0,20,100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y)
plt.plot(y,z)
plt.title("sine and Cosine Wav")
```

```python
# using scatter
x = [5,7,8,7,2,17,2,9]
y = [99,86,87,88,111,22,34,12]
plt.scatter(x,y)
plt.title('Sample Scatter Plot', fontsize= 16)
```

```python
x = ["Engineer", "Arts", "Science", "Business"]
y = [120,80,150,100]
plt.bar(x,y)
plt.title("Highest Major Count")
plt.xlabel("No of Students")
plt.ylabel("Major Name")
```

```python
x = np.linspace(0,5,100)
y1 = x
y2 = x**2
y3 = x**3

plt.plot(x, y1, label= 'y=x', linestyle = '-',color = "blue")
plt.plot(x, y2, label= 'y=x^2', linestyle = '--',color = "red")
plt.plot(x, y3, label= 'y=x^3', linestyle = '-.',color = "green")
plt.legend()
```
